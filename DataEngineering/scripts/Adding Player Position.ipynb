Adding player position column to the existing feature matrix

```python
import numpy as np
import pandas as pd

players = pd.read_csv('players.csv')
feature_matrix = pd.read_csv('feature_matrix.csv')
feature_matrix = feature_matrix.drop(columns = ['Unnamed: 0'], axis = 1)
```

```python
feature_matrix
```

```python
feature_matrix = feature_matrix.merge(players[['id', 'primaryPositioncode', 'primaryPositionname', 'primaryPositiontype']], how = 'left', left_on = 'player_id', right_on = 'id')
```

```python
feature_matrix = feature_matrix.drop(columns = ['id'])
feature_matrix.to_csv('feature_matrix_w_position.csv')
```
